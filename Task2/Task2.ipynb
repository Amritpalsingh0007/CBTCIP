```python
Uemp_India_df = Uemp_India_df.dropna()
Uemp_India_df.info()
```

- **Uemp_India_df can be used to determine trends according to rural and urban area Overall or state wise from May 2019 to June 2020.**
- **Uemp_11_2020_df can be used to determine trends according to state wise or overall from jan 2020 to Oct 2020.**

```python
Uemp_India_df.columns = Uemp_India_df.columns.str.strip()
Uemp_11_2020_df.columns = Uemp_11_2020_df.columns.str.strip()
Uemp_India_df.Date = pd.to_datetime(Uemp_India_df['Date']).dt.date
Uemp_11_2020_df.Date = pd.to_datetime(Uemp_11_2020_df['Date']).dt.date
```

```python
Uemp_11_2020_df.info()
```

```python
Uemp_India_df.info()
```

```python
plt.figure(figsize=(10,10))
plt.plot(Uemp_11_2020_df.Date.unique(),Uemp_11_2020_df.groupby(by = 'Date')['Estimated Unemployment Rate (%)'].mean())
plt.xlabel("Months",fontsize = 12, fontweight = 'bold')
plt.ylabel("Estimated Unemployment Rate (%)",fontsize = 12, fontweight = 'bold')
plt.title("Monthly Estimated Unemployment Rate Of India From 01-2020 To 10-2020",fontsize = 13, fontweight = 'bold')
```

```python
plt.figure(figsize=(10,10))
plt.plot(Uemp_11_2020_df.Date.unique(),Uemp_11_2020_df.groupby(by = 'Date')['Estimated Labour Participation Rate (%)'].mean())
plt.xlabel("Months",fontsize = 12, fontweight = 'bold')
plt.ylabel("stimated Labour Participation Rate (%)",fontsize = 12, fontweight = 'bold')
plt.title("Monthly Estimated Unemployment Rate Of India From 01-2020 To 10-2020",fontsize = 13, fontweight = 'bold')
```

- **From above time series graph we can observe that in month of April and May around 10% to 12% rise in the unemployment rate which means 10% to 12% labour force has stop looking for jobs or they had been fired or had resign from there job. vice versa for the labour paticipation rate.**

```python
data = (Uemp_India_df.groupby(by = ['Date','Area'])['Estimated Unemployment Rate (%)'].mean()).head(28).reset_index()

plt.figure(figsize =(10,10))
sns.lineplot(x='Date',y='Estimated Unemployment Rate (%)',hue = 'Area',data = data)
plt.title("Monthly Estimated Unemployment Rate Of India by Region From 01-2020 To 10-2020",fontsize = 13, fontweight = 'bold')
```

```python
plt.figure(figsize = (10,10))
plt.barh(y =Uemp_11_2020_df.Region.unique(),width = Uemp_11_2020_df.groupby(by = 'Region')['Estimated Unemployment Rate (%)'].mean())
```

```python
plt.figure(figsize = (10,10))
plt.barh(y =Uemp_11_2020_df.Region.unique(),width = Uemp_11_2020_df.groupby(by = 'Region')['Estimated Labour Participation Rate (%)'].mean())
plt.xlabel("Estimated Labour Participation Rate (%)", fontweight = 'bold', fontsize = 12)
plt.ylabel("Region", fontweight = 'bold', fontsize = 12)
plt.title("Avg State Labour Participation Rate(%)", fontsize = 14, fontweight = 'bold')
```

```python
data = Uemp_India_df.groupby(by = ['Region','Area'])['Estimated Unemployment Rate (%)'].mean().reset_index()

plt.figure(figsize = (10, 10))
sns.barplot(y = 'Region', x = 'Estimated Unemployment Rate (%)', hue = 'Area', data = data, orient = 'h')
plt.title("States Rural and urban Unemployment rate (%)",fontweight = 'bold')
plt.xlabel("Estimated Unemployment Rate (%)",fontsize = 12,fontweight = 'bold')
plt.ylabel("Region",fontsize = 12,fontweight = 'bold')
```

```python
data = Uemp_India_df.groupby(by = ['Region','Area'])['Estimated Labour Participation Rate (%)'].mean().reset_index()

plt.figure(figsize = (10, 10))
sns.barplot(y = 'Region', x = 'Estimated Labour Participation Rate (%)', hue = 'Area', data = data, orient = 'h')
plt.title("States Rural and urban Labour Paticipaiton rate (%)",fontweight = 'bold')
plt.xlabel("Estimated Labour Participation Rate (%)",fontsize = 12,fontweight = 'bold')
plt.ylabel("Region",fontsize = 12,fontweight = 'bold')
```

- **AS we can the above graph there is heigher unemployment rate in urban area as compare to the rural area and there is are two main reason that is populations difference and coronavirus has more victims in urban area due to dense population.**
- **From the bar graphs Haryana and tripura have highest avg Unemployment rate among all states.**
- **From the above graphs Meghalaya, Tripura and Telangana have the highest avg Labour Participation Rate (%). Here  we can say that in Tripura more people are  looking for jobs in 2020**
